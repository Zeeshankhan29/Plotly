```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import plotly.express as px
```

```python
df = pd.read_csv('daily-min-temperatures.csv')
df
```

```python
px.ecdf(data_frame=df,x='Temp')
```

```python
px.line(data_frame=df,x='Date',y='Temp',title='Analysis')
```

```python
px.box(data_frame=df,x='Temp')
```

```python
px.violin(data_frame=df,x='Temp')
```

```python
px.histogram(data_frame=df,x='Temp')
```

```python
px.scatter(data_frame=df,x='Temp')
```

```python
df1 = sn.load_dataset('iris')
df1
```

```python
px.scatter(data_frame=df1,x='sepal_length',y='sepal_width',size='petal_length',color='species',hover_data=['petal_width']
          ,marginal_x="histogram")
```

```python
px.scatter(data_frame=df1, x="sepal_length", y="petal_length", color="species", facet_col="species")
```

```python
px.scatter_matrix(data_frame=df1, dimensions=["sepal_width", "sepal_length", "petal_width", "petal_length"], color="species")
```

```python
px.scatter_3d(data_frame=df1, x='sepal_width', y='sepal_length', z='petal_width',
              color='species')
```
